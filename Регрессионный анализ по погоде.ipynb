```python
temp_mean = df2_no_outliers.groupby(df2_no_outliers['Date'].dt.week)['Temperature'].mean()
bikes_sum =  df2_no_outliers.groupby(df2_no_outliers['Date'].dt.week)['Rental_count'].sum()
bikes_sum
```

```python
bk_week = pd.concat([temp_mean, bikes_sum], axis = 1)
bk_week
```

```python
bk_week.corr()
```

```python
bk_week.plot('Temperature', 'Rental_count', kind = 'scatter')
```

```python
from statsmodels.formula.api import ols
```

```python
model = ols("Rental_count ~ Temperature", bk_week).fit() 
print(model.summary())#выдача информации
```

Значение Jarque-Bera также указывает на то, что модель не является абсолютно нормально распределенной (вероятность ошибки составляет 0.167). Скос данных равен 0.640, что указывает на то, что данные смещены вправо. Куртозис равен 2.878, что указывает на то, что данные имеют более плоское распределение, чем нормальное. Значение Cond. No. указывает на то, что в модели может быть присутствовать мультиколлинеарность (высокая корреляция между предикторами).

```python
temp_mean = df2_no_outliers.groupby(df2_no_outliers['Date'].dt.week)['Wind_speed'].mean()
bikes_sum =  df2_no_outliers.groupby(df2_no_outliers['Date'].dt.week)['Rental_count'].sum()
bikes_sum
```

```python

from statsmodels.formula.api import ols
model = ols("Rental_count ~ Wind_speed", bk_week).fit() 
print(model.summary())#выдача информации
```

Вывод: Данные Omnibus являются мерой нормальности распределения остатков модели. Значение Omnibus равно 3.793, а вероятность Prob(Omnibus) равна 0.150. Чем ближе значение Omnibus к нулю, тем более нормально распределены остатки. В данном случае значение Omnibus не сильно отличается от нуля, что говорит о том, что остатки модели достаточно близки к нормальному распределению.

Значение Durbin-Watson равно 0.619 и является мерой автокорреляции остатков модели. Значение Durbin-Watson находится в диапазоне от 0 до 4, и чем ближе оно к 2, тем меньше автокорреляция. В данном случае значение Durbin-Watson меньше 2, что может указывать на наличие положительной автокорреляции в остатках модели.

Значение Jarque-Bera (JB) равно 3.657, а вероятность Prob(JB) равна 0.161. Jarque-Bera тест используется для проверки нормальности распределения остатков модели. Если значение JB мало и вероятность Prob(JB) большая, то можно сделать вывод, что остатки распределены нормально. В данном случае значение JB не сильно отличается от нуля, и вероятность Prob(JB) достаточно высока, что говорит о том, что остатки модели распределены нормально.

Kurtosis (эксцесс) равен 2.565 и является мерой "толщины" распределения остатков. Значение эксцесса равное 0 соответствует нормальному распределению. В данном случае значение эксцесса больше 0, что может указывать на наличие "толстых" хвостов в распределении остатков.

Cond. No. (условное число) равно 12.3 и является мерой мультиколлинеарности в регрессионной модели. Чем больше значение Cond. No., тем более выражена мультиколлинеарность. В данном случае значение Cond. No. равно 12.3, что может указывать на наличие мультиколлинеарности в модели, однако для точного вывода требуется больше информации о данных и модели.
